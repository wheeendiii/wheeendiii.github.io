```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ipywidgets 
from ipywidgets import interact
import cufflinks as cf
import plotly.graph_objs as go
```

Allow users to change the x and y fields on scatter plot


```python
# import the UFO dataset 
ufo = pd.read_csv("ufo-scrubbed-geocoded-time-standardized-00.csv",
                  names = ["date", "city", "state", "country",
                          "shape", "duration_seconds", "duration",
                          "comment", "report_date", "latitude", "longitude"],
                  parse_dates = ["date", "report_date"])
ufo
```

```python
# Fill NaN 
ufo = ufo.fillna({'country':'unknown', 'state': 'unknown', 'shape':'unknown'})
ufo
```

Select random 1000 samples.

```python
nsamples = 1000
downSampleMask = np.random.choice(range(len(ufo)-1), nsamples, replace=False)
ufoDS = ufo.loc[downSampleMask]
```

### Q1: Allow users to change axis fields and pick colormap

```python
@ipywidgets.interact(colormap = plt.colormaps(),
                     x = ['state','country','shape','report_date','longitude','latitude'], 
                     y = ['duration_seconds','longitude','latitude'])
def choose_fields(x , y,colormap):
    plt.figure(figsize=(13,10))
    plt.scatter(ufoDS[x], ufoDS[y], 
                c=np.log10(ufoDS['duration_seconds']),cmap = colormap)
    plt.xticks(rotation = 90)
    plt.colorbar()
```

#### Do you want the user to be able to pick any 2 columns?  Or only certain columns?  Why or why not?

I do not want users to pick any 2 columns because some variables like 'comments'cannot be graphed. Numerical and categorical data are prefered in this graph. Users are able to choose either catehorical or numerial column in x axis and numerial columns in y axis.  However, users can't choose 'city' column in this interactive graph because due to the large number of cities, the visualization will not be able to display city names clearly.

### Q2: Build widget to dispay aggregated values of columns

```python
@ipywidgets.interact(x = ['state','city','country','shape','report_date','duration_seconds','longitude','latitude'],
                     y = ['duration_seconds','longitude','latitude'])

def binned_graph(x,y):
    initial_bin_width = 10
    if(ufoDS[x].dtype == '<M8[ns]'or ufoDS[x].dtype =='O'):
        plot = ufoDS.groupby(x)[y].sum()
        plot.plot(figsize=(13,10))
    
    else:
        figure_widget = go.FigureWidget(
        data=[go.Histogram(x=ufoDS[x], xbins={"size": initial_bin_width})]
    )

        bin_slider = ipywidgets.FloatSlider(
            value=initial_bin_width,
            min=1,
            max=30,
            step=1,
            description="Bin width:",
            readout_format=".0f",  # display as integer
        )

        histogram_object = figure_widget.data[0]

        def set_bin_size(change):
            histogram_object.xbins = {"size": change["new"]}

        bin_slider.observe(set_bin_size, names="value")

        output_widget = ipywidgets.VBox([figure_widget, bin_slider])
        return output_widget
```
